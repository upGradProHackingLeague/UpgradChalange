This script will be used for downloading required files from the pro kabadi site

```python
import numpy as np
import json
import requests
import shutil
import os
import urllib
```

```python
# Team Seasion information 

def GetSeassionID(i):
    x=i
    if i==5:
        x=8
    if i==6:
        x=26
    if i==7:
        x=49
    
    return x

def SaveToDesk(url,filename):
    if not os.path.exists(os.path.dirname(filename)):
        try:
            os.makedirs(os.path.dirname(filename))
        except OSError as exc: # Guard against race condition
            if exc.errno != errno.EEXIST:
                raise
    urllib.request.urlretrieve (url, filename)
    
def data_download_process(_folder,id):
    for i in range(1,8):
        _session_id = GetSeassionID(i)
        _url = f"https://www.prokabaddi.com/sifeeds/kabaddi/static/json/1_" + str(_session_id)+f"_" +str(id)+f"_stats.json";
        _filename = _folder +f"/Season_"+str(i)+".json"
        print(_url)
        SaveToDesk(_url,_filename)
        
```
